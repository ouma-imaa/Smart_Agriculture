```python
import zipfile
import os
from ultralytics import YOLO
from PIL import Image as PILImage
from IPython.display import Image, display
```

## **<font color="blue">Extraction du fichier contenant les données</font>**

```python
# Chemin du fichier zip :
chemin_zip = "/content/cvv.zip"

# Chemin du dossier de destination :
chemin_destination = "/content/dataset"

# Créer le dossier de destination s'il n'existe pas
os.makedirs(chemin_destination, exist_ok=True)

# Extraire le contenu du fichier zip dans le dossier de destination
with zipfile.ZipFile(chemin_zip, 'r') as zip_ref:
    zip_ref.extractall(chemin_destination)

print("Extraction terminée.")

```

## **<font color="blue">Entrainement de modèle YOLOV8</font>**

```python
# Load a model
model = YOLO("yolov8n.yaml")

# Train the model
results = model.train(data = "/content/dataset/cvv/data.yaml", epochs=100)
```

### **<font color="blue">F1 curve</font>**

```python
# Chemin vers l'image F1_curve.png
image_path = '/content/runs/detect/train2/F1_curve.png'

# Charger l'image avec Pillow
original_image = PILImage.open(image_path)

# Redimensionner l'image (par exemple, à la taille 50%)
width, height = original_image.size
new_size = (int(width * 0.4), int(height * 0.4))
resized_image = original_image.resize(new_size)

# Enregistrer l'image redimensionnée dans un nouveau fichier
resized_path = '/content/runs/detect/F1_curve_resized.png'
resized_image.save(resized_path)

# Afficher l'image redimensionnée
display(Image(filename=resized_path))
```

**--- Les scores F1 pour chaque catégorie dans notre ensemble de données d'animaux nuisibles étaient initialement faibles avec un seuil de confiance bas.
